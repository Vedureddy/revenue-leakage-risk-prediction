# PHASE 3 ‚Äî DATA INTEGRATION

# STEP 1 ‚Äî Aggregate ORDER ITEMS (pricing & shipping)

```python
# ----------------------------------------------------------
# ORDER ITEMS TABLE has multiple rows per order (one per item)
# We convert it to ORDER LEVEL (one row per order)
# ----------------------------------------------------------

items_agg = items.groupby("order_id").agg(

    total_item_value=("price", "sum"),  
    # Sum of all item prices in an order ‚Üí total order value

    total_freight_value=("freight_value", "sum"),  
    # Total shipping cost for the order

    num_items=("price", "count"),  
    # Number of items purchased in the order

    num_sellers=("seller_id", "nunique")  
    # Number of unique sellers involved in the order

).reset_index()   # Convert grouped result back to normal table

```

```python
# ----------------------------------------------------------
# PAYMENTS TABLE may contain multiple payment records per order
# Example: split payments, multiple methods, installments
# Convert to one row per order
# ----------------------------------------------------------

payments_agg = payments.groupby("order_id").agg(

    total_payment=("payment_value", "sum"),  
    # Total amount paid for the order

    max_installments=("payment_installments", "max")  
    # Highest number of installments used

).reset_index()
```

```python
# ----------------------------------------------------------
# PAYMENTS TABLE may contain multiple payment records per order
# Example: split payments, multiple methods, installments
# Convert to one row per order
# ----------------------------------------------------------

payments_agg = payments.groupby("order_id").agg(

    total_payment=("payment_value", "sum"),  
    # Total amount paid for the order

    max_installments=("payment_installments", "max")  
    # Highest number of installments used

).reset_index()
```

```python
# ----------------------------------------------------------
# REVIEWS TABLE gives rating per order
# Usually one review per order, but aggregate for safety
# ----------------------------------------------------------

reviews_agg = reviews.groupby("order_id").agg(

    avg_review_score=("review_score", "mean")  
    # Average rating for the order (1‚Äì5 scale)

).reset_index()
```

# üîó STEP 4 ‚Äî Merge All Data into ONE Dataset

```python
# ----------------------------------------------------------
# Add customer location information to each order
# Join using customer_id
# how="left" ‚Üí keep all orders even if customer info missing
# ----------------------------------------------------------

df = orders.merge(customers, on="customer_id", how="left")
```

```python
# ----------------------------------------------------------
# Add aggregated pricing & shipping features
# Join using order_id
# ----------------------------------------------------------

df = df.merge(items_agg, on="order_id", how="left")
```

```python
# ----------------------------------------------------------
# Add payment behavior features
# ----------------------------------------------------------

df = df.merge(payments_agg, on="order_id", how="left")
```

```python
# ----------------------------------------------------------
# Add customer satisfaction (review score)
# ----------------------------------------------------------

df = df.merge(reviews_agg, on="order_id", how="left")
```

# PHASE 5 ‚Äî Target Creation

```python
# ----------------------------------------------------------
# Create Revenue Leakage Risk label
# 1 = High Risk
# 0 = Low Risk
# ----------------------------------------------------------

df["leakage_risk"] = (
    
    (df["is_late"] == 1) |                     # Delivered late
    
    (df["freight_ratio"] > 0.5) |              # Very high shipping cost
    
    (df["avg_review_score"] <= 2) |            # Poor rating
    
    (df["not_delivered"] == 1) |               # Cancelled / not delivered
    
    (df["max_installments"] >= 6)              # Heavy installment usage
    
).astype(int)
```

# ü§ñ PHASE 7 ‚Äî MODELING

```python
# ----------------------------------------------------------
# STEP 1 ‚Äî Split data into training and testing sets
# ----------------------------------------------------------

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,      # 20% for testing
    random_state=42,    # Reproducible results
    stratify=y          # Preserve class balance
)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
# ----------------------------------------------------------
# STEP 2 ‚Äî Train Logistic Regression
# ----------------------------------------------------------

from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(max_iter=1000)

log_model.fit(X_train, y_train)

print("Logistic Regression trained ‚úî")
```

```python
# ----------------------------------------------------------
# STEP 3 ‚Äî Train Random Forest
# ----------------------------------------------------------

from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

rf_model.fit(X_train, y_train)

print("Random Forest trained ‚úî")
```

```python
# ----------------------------------------------------------
# STEP 4 ‚Äî Train Gradient Boosting
# ----------------------------------------------------------

from sklearn.ensemble import GradientBoostingClassifier

gb_model = GradientBoostingClassifier(random_state=42)

gb_model.fit(X_train, y_train)

print("Gradient Boosting trained ‚úî")
```

# PHASE 8 ‚Äî EVALUATION (Correct Order)

```python
# ----------------------------------------------------------
# STEP 1 ‚Äî Predict leakage risk on test data
# Each model predicts labels (0 = Low risk, 1 = High risk)
# ----------------------------------------------------------

y_pred_log = log_model.predict(X_test)   # Logistic Regression predictions
y_pred_rf  = rf_model.predict(X_test)    # Random Forest predictions
y_pred_gb  = gb_model.predict(X_test)    # Gradient Boosting predictions
```

```python
# ----------------------------------------------------------
# STEP 2 ‚Äî Calculate accuracy
# Accuracy = % of correctly predicted orders
# ----------------------------------------------------------

from sklearn.metrics import accuracy_score

print("Logistic Accuracy:", accuracy_score(y_test, y_pred_log))
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Gradient Boosting Accuracy:", accuracy_score(y_test, y_pred_gb))
```

```python
# ----------------------------------------------------------
# STEP 3 ‚Äî Detailed classification metrics
# Precision = How many predicted risks are truly risks
# Recall    = How many actual risks were detected
# F1-score  = Balance between precision and recall
# ----------------------------------------------------------

from sklearn.metrics import classification_report

print("\n--- Logistic Regression ---")
print(classification_report(y_test, y_pred_log))

print("\n--- Random Forest ---")
print(classification_report(y_test, y_pred_rf))

print("\n--- Gradient Boosting ---")
print(classification_report(y_test, y_pred_gb))
```

```python
# ----------------------------------------------------------
# STEP 4 ‚Äî Confusion Matrix
# Shows exact counts of prediction outcomes
# ----------------------------------------------------------

from sklearn.metrics import confusion_matrix

print("\nLogistic Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_log))

print("\nRandom Forest Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf))

print("\nGradient Boosting Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_gb))
```
