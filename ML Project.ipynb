# PHASE 3 — DATA INTEGRATION

```python


items_agg = items.groupby("order_id").agg(

    total_item_value=("price", "sum"),  


    total_freight_value=("freight_value", "sum"),  


    num_items=("price", "count"),  
 

    num_sellers=("seller_id", "nunique")  


).reset_index() 
```

```python


payments_agg = payments.groupby("order_id").agg(

    total_payment=("payment_value", "sum"),  


    max_installments=("payment_installments", "max")  


).reset_index()
```

```python
payments_agg = payments.groupby("order_id").agg(

    total_payment=("payment_value", "sum"),  
 

    max_installments=("payment_installments", "max")  


).reset_index()
```

```python


reviews_agg = reviews.groupby("order_id").agg(

    avg_review_score=("review_score", "mean")  
).reset_index()
```

```python


df = orders.merge(customers, on="customer_id", how="left")
```

```python


df = df.merge(items_agg, on="order_id", how="left")
```

```python

df = df.merge(payments_agg, on="order_id", how="left")
```

```python

df = df.merge(reviews_agg, on="order_id", how="left")
```

```python
df["freight_ratio"] = (
    df["total_freight_value"] / df["total_item_value"]
)

print("order id and percent of price for shipping( ex: 0.2 = 20% for shipping or delivery cost)")
print(df["freight_ratio"].head())
```

```python
df["installment_per_value"] = (
    df["max_installments"] / df["total_order_value"]
)

print("\nInstallment per order id  created:")
print(df["installment_per_value"].head())
```

```python
df["is_SP_customer"] = (df["customer_state"] == "SP").astype(int)

print("\nSP Customer Flag created:")
print(df["is_SP_customer"].head())
```

```python
df["avg_review_score"] = df["avg_review_score"].fillna(3)

print("\nReview Score after filling missing values:")
print(df["avg_review_score"].head())
```

```python
df["is_late"] = (df["delay_days"] > 0).astype(int)

print("\nLate Delivery Flag created:")
print(df["is_late"].head())
print("1 → Delivered late  0 → On time or early")
```

```python
print(df["is_late"].unique())
```

```python
df["not_delivered"] = (
    df["order_status"] != "delivered"
).astype(int)

print("\nNot Delivered Flag created:")
print(df["not_delivered"].head())
print(df["not_delivered"].unique())
print("If order_status != 'delivered' → 1 If order_status == 'delivered' → 0")
```

# PHASE 5 — Target Creation

```python


df["leakage_risk"] = (
    
    (df["is_late"] == 1) |                     # Delivered late
    
    (df["freight_ratio"] > 0.5) |              # Very high shipping cost
    
    (df["avg_review_score"] <= 2) |            # Poor rating
    
    (df["not_delivered"] == 1) |               # Cancelled / not delivered
    
    (df["max_installments"] >= 6)              # Heavy installment usage
    
).astype(int)
```

```python
num_cols = df.select_dtypes(include=["int64", "float64"]).columns
df[num_cols] = df[num_cols].fillna(0)
```

```python
cat_cols = df.select_dtypes(include=["object"]).columns
df[cat_cols] = df[cat_cols].fillna("Unknown")
```

```python
cols_to_drop = ["order_id", "customer_id", "seller_id", "product_id"]

existing = [c for c in cols_to_drop if c in df.columns]

df = df.drop(columns=existing)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

features_to_scale = df.select_dtypes(include=["int64", "float64"]).columns.drop("leakage_risk")

df[features_to_scale] = scaler.fit_transform(df[features_to_scale])
```

```python
print(df.shape)
print(df.dtypes.value_counts())
print(df.isnull().sum().sum())
```

```python
y = df["leakage_risk"]
```

```python
X = df.drop(columns=["leakage_risk"])
```

# PHASE 7 — MODELING

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,      # 20% for testing
    random_state=42,    # Reproducible results
    stratify=y          # Preserve class balance
)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(max_iter=1000)

log_model.fit(X_train, y_train)

print("Logistic Regression trained ✔")
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

rf_model.fit(X_train, y_train)

print("Random Forest trained ✔")
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb_model = GradientBoostingClassifier(random_state=42)

gb_model.fit(X_train, y_train)

print("Gradient Boosting trained ✔")
```

# PHASE 8 — EVALUATION

```python
y_pred_log = log_model.predict(X_test)   # Logistic Regression predictions
y_pred_rf  = rf_model.predict(X_test)    # Random Forest predictions
y_pred_gb  = gb_model.predict(X_test)    # Gradient Boosting predictions
```

```python
from sklearn.metrics import accuracy_score

print("Logistic Accuracy:", accuracy_score(y_test, y_pred_log))
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Gradient Boosting Accuracy:", accuracy_score(y_test, y_pred_gb))
```

```python
from sklearn.metrics import classification_report

print("\n--- Logistic Regression ---")
print(classification_report(y_test, y_pred_log))

print("\n--- Random Forest ---")
print(classification_report(y_test, y_pred_rf))

print("\n--- Gradient Boosting ---")
print(classification_report(y_test, y_pred_gb))
```

```python
from sklearn.metrics import confusion_matrix

print("\nLogistic Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_log))

print("\nRandom Forest Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf))

print("\nGradient Boosting Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_gb))
```
